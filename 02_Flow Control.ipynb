```python
#1.] Write a python program to read three numbers and if any two 
#variables are equal, print that number
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

if a == b:
    print("Equal number is:", a)
elif a == c:
    print("Equal number is:", a)
elif b == c:
    print("Equal number is:", b)
else:
    print("No two numbers are equal")
```

```python
#4.] Write a python program to read three numbers and print them in 
#ascending order (without using any function)
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
c = int(input("Enter third number: "))

# Compare and print in ascending order
if a <= b and a <= c:        # a is smallest
    if b <= c:
        print(a, b, c)
    else:
        print(a, c, b)

elif b <= a and b <= c:      # b is smallest
    if a <= c:
        print(b, a, c)
    else:
        print(b, c, a)

else:                        # c is smallest
    if a <= b:
        print(c, a, b)
    else:
        print(c, b, a)
```

```python
#5.] Write a python program to read four numbers (representing the 
#four octets of an IP) and check whether they all are in the range 
#between 0 to 255 (networking mai padha tha)
a = int(input("Enter 1st octet: "))
b = int(input("Enter 2nd octet: "))
c = int(input("Enter 3rd octet: "))
d = int(input("Enter 4th octet: "))

if (a >= 0 and a <= 255) and \
   (b >= 0 and b <= 255) and \
   (c >= 0 and c <= 255) and \
   (d >= 0 and d <= 255):
    print("Valid IP Address")
else:
    print("Invalid IP Address")
```

```python
#10] Write a python program to read three numbers (a,b,c) and check 
#how many numbers between ‘a’ and ‘b’ are divisible by ‘c’
a = int(input("Enter a: "))
b = int(input("Enter b: "))
c = int(input("Enter c: "))

i = a
result = 0  # final count stored here implicitly using result
while i <= b:
    if i % c == 0:
        result += 1  # flow control + loop वापरून increment
    i += 1

print("Numbers divisible by", c, "between", a, "and", b, "are:", result)
```

```python
#12.] Write a python program to print the factorial of a given number

num = int(input("Enter a number: "))

factorial = 1
i = 1

while i <= num:
    factorial *= i
    i += 1

print("Factorial of", num, "is", factorial)
```

```python
#13.] Write a python program to print the first 18 terms of a Fibonacci series
a = 0
b = 1
i = 1  # term index

while i <= 18:
    print(a, end="  ")
    c = a + b
    a = b
    b = c
    i += 1
```

```python
#14.] Write a python program to read a number and print a right triangle using "*"
#Eg : 
#Input : 5
#----------Output--------- 
#*
#* *
#* * *
#* * * *
#* * * * *
n = int(input("Enter a number: "))

for i in range(1, n + 1):
    for j in range(1, i + 1):
        print("*", end=" ")
    print()  # new line after each row
```
