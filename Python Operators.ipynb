```python
#3.] Basic salary is input to the keyboard. 
#Write a program to print a gross salary 
#if da is 58% of basic, ta is 30% of basic and hra is 38% 
#of basic.

basic = float(input("Enter Basic Salary: "))

da = 0.58 * basic
ta = 0.30 * basic
hra = 0.38 * basic

gross = basic + da + ta + hra

print("Gross Salary =", gross)
```

```python
#10.] Consider a room having one door and two windows 
# of the same size. Accept dimensions of the room, 
#door and window. Print the area to be painted 
#(interior walls) and area to be whitewashed (roof).

# Room dimensions
L = float(input("Enter room length: "))
B = float(input("Enter room breadth: "))
H = float(input("Enter room height: "))

# Door dimensions
Dw = float(input("Enter door width: "))
Dh = float(input("Enter door height: "))

# Window dimensions
Ww = float(input("Enter window width: "))
Wh = float(input("Enter window height: "))

# Total wall area (2*(L*H + B*H))
wall_area = 2 * (L*H + B*H)

# Door area
door_area = Dw * Dh

# Window area (2 windows)
window_area = 2 * (Ww * Wh)

# Paint area = wall area - door - windows
paint_area = wall_area - door_area - window_area

# Roof area = L * B
whitewash_area = L * B

print("Area to be painted (walls) =", paint_area)
print("Area to be whitewashed (roof) =", whitewash_area)
```

```python
#11.] Write a program to convert specified days into years, 
#weeks and days. Note: Ignore leap year. assumes that all 
#months have 30 days and all years have 365 days
#i/p : Number of days : 1329
#Expected Output :  Years: 3    Weeks: 33      Days: 3

total_days = int(input("Enter number of days: "))

years = total_days // 365
remaining_days = total_days % 365

weeks = remaining_days // 7

days = remaining_days % 7

print("Years:", years ) 
print("Weeks:", weeks)
print("Days:", days )
```

```python
#15.] N students take K apples and distribute them among 
#each other evenly. The remaining (the undivisible) part 
#remains in the basket. How many apples will each single 
#student get? How many apples will remain in the basket? 
#The program reads the numbers N and K.

N = int(input("Enter number of students: "))
K = int(input("Enter total number of apples: "))

apples_per_student = K // N

remaining_apples = K % N

print("Each student gets =", apples_per_student, "apples")
print("Remaining apples in basket =", remaining_apples)
```

```python
#16.] A timestamp is three numbers: a number of hours, 
#minutes and seconds. Given two timestamps, 
#calculate how many seconds is between them. 
#The moment of the first timestamp occurred before the 
#moment of the second timestamp.


# Input first timestamp
h1 = int(input("Enter hours of first timestamp: "))
m1 = int(input("Enter minutes of first timestamp: "))
s1 = int(input("Enter seconds of first timestamp: "))

# Input second timestamp
h2 = int(input("Enter hours of second timestamp: "))
m2 = int(input("Enter minutes of second timestamp: "))
s2 = int(input("Enter seconds of second timestamp: "))

# Convert to total seconds 
total_seconds1 = h1*3600 + m1*60 + s1
total_seconds2 = h2*3600 + m2*60 + s2

difference = total_seconds2 - total_seconds1

print("Difference between timestamps in seconds =", difference)
```

```python
#17.] A school decided to replace the desks in three 
#classrooms. Each desk sits two students. Given the 
#number of students in each class, print the smallest 
##possible number of desks that can be purchased. 
#The program should read three integers: 
#the number of students in each of the three classes, a, b and c respectively. 
#eg: The first class has 20 students and thus needs 10 desks.
#The second class has 21 students, so they can get by with no fewer than 11 desks.
#11 desks is also enough for the third class of 22 students. So we need 32 desks 
#in total.

a = int(input("Enter number of students in class 1: "))
b = int(input("Enter number of students in class 2: "))
c = int(input("Enter number of students in class 3: "))

# Desks needed for each class
desks1 = (a + 1) // 2
desks2 = (b + 1) // 2
desks3 = (c + 1) // 2

# Total desks
total_desks = desks1 + desks2 + desks3

print("Minimum number of desks needed =", total_desks)
```

```python
#20.] A car can cover distance of N kilometers per day. How many days will it take to cover a route of length 
#M kilometers? The program gets two numbers: N and M.

# Input
N = int(input("Distance per day (N km): "))
M = int(input("Total route length (M km): "))

# Days calculation (ceiling division using operators only)
days = (M + N - 1) // N

print("Number of days required =", days)

```

```python
#22.] A cake costs A rupees and B paise. Determine, 
#how many rupees and paise cents should one pay for N cakes. A program gets three numbers: A, B, N. 
#It should print two numbers: total cost in rupees and paise. 

A = int(input("Rupees per cake: "))
B = int(input("Paise per cake: "))
N = int(input("Number of cakes: "))

total_paise = N * (A*100 + B)
total_rupees = total_paise // 100
remaining_paise = total_paise % 100

print("Total cost =", total_rupees, "rupees", remaining_paise, "paise")
```
